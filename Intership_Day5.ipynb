```python
import numpy as np
products=["Laptop","Mobile","Tablet","HeadPhone","Smartwatch"]
sales=np.array([120,180,80,150,100])
print("Total Sales:",np.sum(sales))
print("Average Sales:",np.mean(sales))
print("Highest sales",np.max(sales))
print("Lowewst sales",np.min(sales))
best_index=np.argmax(sales)
print("Best selling products:",products[best_index])
print("\n Product sales report")
for i in range(len(products)):
    print(products[i],":",sales[i])
```

```python
import pandas as pd
employees={
    "Emp_ID":[101,102,103,104,105],
    "Name":["Arun","Priya","Rahul","Divya","Kumar"],
    "Department":["IT","HR","IT","Finance","HR"],
    "Salary":[50000,40000,50000,55000,42000],
    "Experience":[2,3,5,6,4]
}
df=pd.DataFrame(employees)
print("Employee Dataset")
print(df)
print("\nAverage Salary")
print(df["Salary"].mean())
print("\nHighest Salary")
print(df["Salary"].max())
print("\nEmployees in HR")
print(df[df["Department"]=="HR"])
print("\nEmployees with experience>3")
```
